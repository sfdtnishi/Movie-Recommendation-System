```python
#importing dependencies
import pandas as pd  #analyse data nd manipulation of data
import numpy as np #arrays
import difflib #if in case movie name is misspelled by user, it will choose the closest matching movie from dataset.
from sklearn.feature_extraction.text import TfidfVectorizer   #to convert textual data into numnerical values
from sklearn.metrics.pairwise import cosine_similarity #to know the similarity score for movie matching
```

```python
#data collection and pre-processing
#loading the data from csv file to pandas dataframe
movie_data = pd.read_csv("movies.csv")  #movie_data is the dataframe here.
```

```python
#printing first 5 rows of dataframe using head
movie_data.head()
```

```python
#to find number of rows and columns in data frame
movie_data.shape
```

```python
#selecting the relevant features for recommendation
selected_features = ['genres','keywords','tagline','cast','director'] #based on our content recommendation system
print(selected_features)
```

```python
#textual data contains null values, so here we're converting these values into null strings.
for feature in selected_features:
    movie_data[feature]=movie_data[feature].fillna('')
```

```python
#combinng all the five selected features
combined_features = movie_data['genres']+' '+movie_data['keywords']+' '+movie_data['tagline']+' '+movie_data['cast']+' '+movie_data['director']
```

```python
print(combined_features)
```

```python
#converting text data into feature vectors
vectorizer = TfidfVectorizer() #here vectorizer is a variable
```

```python
feature_vectors=vectorizer.fit_transform(combined_features.apply(lambda x: np.str_(x)))
```

```python
print(feature_vectors)
```

```python
#getting cosine similarity score using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape) # (movie index, similarity score) : (4803,4803)
```

```python
#getting the movie name from the user
movie_name= input("enter your favourite movie name: ")
```

```python
#creating a list of all the movies given in the dataset
list_of_all_titles= movie_data['title'].tolist() #tolist function is used to create list of selected data
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by the user. 
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles) #comparing movie given by user from the mobies present in the dataset
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title
index_of_the_movie= movie_data[movie_data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))  #emumerate: runs in loop through list also giving the index
print(similarity_score)
```

```python
len(similarity_score) #no. of movies being compared
```

```python
#sorting the movies based on their similarity score
sorted_similarity_movies=sorted(similarity_score, key= lambda x:x[1], reverse=True)
print(sorted_similarity_movies)
```

```python
#printing the name of similar movies based index
print('movies suggested for you: \n')
i=1
for movie in sorted_similarity_movies:
    index=movie[0]
    title_from_index=movie_data[movie_data.index==index]['title'].values[0]
    if (i<30):
        print(i,'. ',title_from_index)
        i+=1
```

```python
#movie recommendation system
movie_name= input("enter your favourite movie name: ")
list_of_all_titles= movie_data['title'].tolist() 
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
close_match=find_close_match[0]
index_of_the_movie= movie_data[movie_data.title==close_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_the_movie]))
sorted_similarity_movies=sorted(similarity_score, key= lambda x:x[1], reverse=True)
print('\n movies suggested for you: \n')
i=1
for movie in sorted_similarity_movies:
    index=movie[0]
    title_from_index=movie_data[movie_data.index==index]['title'].values[0]
    if (i<30):
        print(i,'. ',title_from_index)
        i+=1
```
